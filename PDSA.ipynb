```python
import sys
sys.setrecursionlimit(2**31-1)
```

```python
def quick(L,l,r):
  if r-l<=1:
    return L
  piv=L[l]
  upper=l+1
  lower=l+1
  for i in range(l+1,r):
    if L[i]>piv:
      upper+=1
    else:
      temp=L[i]
      L[i]=L[lower]
      L[lower]=temp
      lower+=1
      upper+=1
  
  L[l]=L[lower-1]
  L[lower-1]=piv
  lower-=1
  quick(L,l,lower)
  quick(L,lower+1,upper)
  return L
```

```python
# Dijkstra example
dedges = [(0,1,10),(0,2,80),(1,2,6),(1,4,20),(2,3,70),(4,5,50),(4,6,5),(5,6,10)]
edges = dedges + [(j,i,w) for (i,j,w) in dedges]
```

```python
def dijkstralist(Wlist,s):
  infinity=1+len(Wlist.keys())*max([d for u in Wlist.keys() for (v,d) in Wlist[u]])
  visited,distance={},{}
  for i in Wlist.keys():
    visited[i],distance[i]=False,infinity
  distance[s]=0
  for u in Wlist.keys():
    nextd=min([distance[v] for v in Wlist.keys() if not visited[v]])
    nextvlist=[v for v in Wlist.keys() if not visited[v] and distance[v]==nextd]
    if nextvlist==[]:
      break
    nextv=min(nextvlist)
    visited[nextv]=True
    for (v,d) in Wlist[nextv]:
      if not visited[v]:
        distance[v]=min(distance[v],distance[nextv]+d)
  return distance
```

## Floyd Warshall

```python
# Floyd Warshall Example
edges = [(0,1,10),(0,7,8),(1,5,2),(2,1,1),(2,3,1),(3,4,3),(4,5,-1),(5,2,-2),(6,1,-4),(6,5,-1),(7,6,1)]
```

```python
size = 8
import numpy as np
W = np.zeros(shape=(size,size,2))
for (i,j,w) in edges:
    W[i,j,0] = 1
    W[i,j,1] = w    
```

```python
def floydWarshall(WMat):
  rows,cols,x=WMat.shape
  infinity=np.max(WMat)*rows*rows+1
  sp=np.zeros(shape=(rows,cols,cols+1))
  for i in range(rows):
    for j in range(cols):
      if WMat[i,j,0]==1:
        sp[i,j,0]=WMat[i,j,1]
      else:
        sp[i,j,0]=infinity
  for k in range(cols):
    for i in range(rows):
      for j in range(cols):
        sp[i,j,k+1]=min(sp[i,j,k],sp[i,k,k]+sp[k,j,k])
  return sp[:,:,cols]
```

```python
def primlist2(WList):
  infinity = 1 + max([d for u in WList.keys() for (v,d) in WList[u]])
  (visited,distance,nbr) = ({},{},{})
  for v in WList.keys():
    (visited[v],distance[v],nbr[v]) = (False,infinity,-1)
  visited[0]=True
  for (v,d) in WList[0]:
    distance[v]=d
    nbr[v]=0
  for i in range(1,len(WList.keys())):
    nextd=min([distance[v] for v in WList.keys() if not visited[v]])
    nextvlist=[v for v in WList.keys() if not visited[v] and distance[v]==nextd]
    if nextvlist==[]:
      break
    nextv=min(nextvlist)
    visited[nextv]=True
    for (v,d) in WList[nextv]:
      if not visited[v]:
        if  distance[v]>d:
          distance[v]=d
          nbr[v]=nextv
  return nbr    
```

## Kruskal

```python
def kruskal(WList):
  edges,component,TE=[],{},[]
  for u in WList.keys():
    edges.extend([(d,u,v) for (v,d) in WList[u]])
    component[u]=u
  edges.sort()
  for (d,u,v) in edges:
    if component[u]!=component[v]:
      TE.append((u,v))
      c=component[u]
      for i in WList.keys():
        if component[i]==c:
          component[i]=component[v]
  return TE
```

```python
def kruskal2(WList):
  members,size,edges,component,TE={},{},[],{},[]
  for u in WList.keys():
    edges.extend([(d,u,v) for (v,d) in WList[u]])
    component[u]=u
    members[u]=[]
    members[u].append(u)
    size[u]=1
  edges.sort()
  for (d,u,v) in edges:
    if component[u]!=component[v]:
      TE.append((u,v))
      c_old=component[u]
      c_new=component[v]
      if size[v]>=size[u]:
        for i in members[u]:
          component[i]=c_new
          members[v].append(i)
          size[v]+=1
      else:
        for i in members[v]:
          component[i]=c_old
          members[u].append(i)
          size[u]+=1
  return TE
```

## Minimizing Lateness

```python
def minimizeLateness(N,jobs):
    sortedL=sorted(jobs, key = lambda x:(x[2],x[1],x[0]))
    optimum=[[] for i in range(N)]
    usedtime={i:0 for i in range(N)}
    i=0
    while i < len(jobs):
        freeline=min(usedtime, key=lambda x:usedtime[x])
        optimum[freeline].append(sortedL[i][0])
        usedtime[freeline]+=sortedL[i][1]
        i+=1
    return optimum
```

## Huffman's Coding

```python
class Node:
  def __init__(self,frequency,symbol = None,left = None,right = None):
    self.frequency = frequency
    self.symbol = symbol
    self.left = left
    self.right = right
```

```python
def Huffman(s):
  huffcode={}
  l=list(s)
  uniq=set(l)
  freq=[]
  for i in uniq:
    freq.append((l.count(i),i))
  nodes=[]
  for nd in sorted(freq):
    nodes.append((nd,Node(nd[0],nd[1])))
  while len(nodes)>1:
    nodes.sort()
    L=nodes[0][1]
    R=nodes[1][1]
    newnode=Node(L.frequency+L.frequency,L.symbol+R.symbol,L,R)
    nodes.pop(0)
    nodes.pop(0)
    #nodes.append(((newnode.frequency,newnode.symbol),newnode))
    nodes.append(((L.frequency + R.frequency, L.symbol + R.symbol),newnode))
  for ch in uniq:
    temp=newnode
    code=''
    while ch!=temp.symbol:
      if ch in temp.left.symbol:
        code+='0'
        temp=temp.left
      else:
        code+='1'
        temp=temp.right
    huffcode[ch]=code
  return huffcode
```

## Longest Common Subwords

```python
def LCW(u,v):
  m,n=len(u),len(v)
  lcw=np.zeros((m+1,n+1))
  max=0
  for i in range(n-1,-1,-1):
    for j in range(m-1,-1,-1):
      if u[i]==v[j]:
        lcw[i,j]=1+lcw[i+1,j+1]
      else:
        lcw[i,j]=0
      if lcw[i,j]>max:
        max=lcw[i,j]
  return max
```

## Longest Common Subsequences

```python
def LCS(u,v):
  m,n=len(u),len(v)
  lcs=np.zeros((m+1,n+1))
  for i in range(n-1,-1,-1):
    for j in range(m-1,-1,-1):
      if u[i]==v[j]:
        lcw[i,j]=1+lcw[i+1,j+1]
      else:
        lcw[i,j]=max(lcw[i,j+1],lcw[i+1,j])
  return lcw[0,0]
```

## Edit Distance

```python
def ED(u,v):
  m,n=len(u),len(v)
  ed=np.zeros((m+1,n+1))
  for i in range(m):
    ed[i,n]=m-i
  for j in range(m):
    ed[m,j]=n-j
  for i in range(m-1,-1,-1):
    for j in range(n-1,-1,-1):
      if u[i]==v[j]:
        ed[i,j]=ed[i+1,j+1]
      else:
        ed[i,j]=1+min(ed[i+1,j],ed[i+1,j+1],ed[i,j+1])
  return ed[0,0]
```

## Matrix Multiplication

```python
def C(dim):
  n=dim.shape[0]
  c=np.zeros((n,n))
  for i in range(n):
    c[i,i]=0
  for diff in range(1,n):
    for i in range(0,n-diff):
      j=i+diff
      c[i,j]=c[i,i]+c[i+1,j]+dim[i][0]*dim[i+1][0]*dim[j][1]
      for k in range(i+1,j+1):
        c[i,j]=min(c[i,j],c[i,k-1]+c[k,j]+dim[i][0]*dim[k][0]*dim[j][1])
  return c[0,n-1]
```
